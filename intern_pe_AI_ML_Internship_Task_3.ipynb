```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ipl_colab.csv')
print(df.info())
print(df.head())

team_wins = df['winner'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values, palette='viridis')
plt.title('Total IPL Wins by Teams')
plt.ylabel('Number of Wins')
plt.xlabel('Teams')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='season', hue='winner', palette='tab10')
plt.title('IPL Winners by Season')
plt.xlabel('Season')
plt.ylabel('Match Wins')
plt.xticks(rotation=45)
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

most_wins = df['winner'].value_counts().idxmax()
print(f"The most consistent winning team is likely: {most_wins}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ipl_colab.csv')

print(df.info())
print(df.head())

team_scores = df.groupby('batting_team')['total'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=team_scores.index, y=team_scores.values, palette='magma')
plt.title('Total Runs Scored by Each Batting Team')
plt.xlabel('Batting Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_team = team_scores.idxmax()
print(f"The team with the highest total runs across matches: {top_team}")
```
